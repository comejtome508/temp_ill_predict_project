```python
import pandas as pd
diag_pd = pd.read_csv("2.국민건강보험공단_진료내역정보_2022(수정).CSV", encoding='cp949')
diag_pd.head()
```

```python
diag_pd["주상병코드"] = diag_pd["주상병코드"].str.slice(0, 3) 
diag_pd["주상병코드"]
```

```python
tem_pd = pd.read_csv("temperatures.csv", encoding='cp949')
tem_pd.head()
```

```python
tem_pd["요양개시일자"] = tem_pd["요양개시일자"].str.slice(1, 11)
tem_pd_2 = tem_pd[["요양개시일자", "최고기온"]]
tem_pd_2.head()
```

```python
inner_join = pd.merge(diag_pd, tem_pd_2, left_on='요양개시일자', right_on='요양개시일자', how='inner')
inner_join.head()
```
